```python
import pandas as pd
from scipy import stats
from statsmodels.formula.api import ols, logit
from statsmodels.stats.anova import anova_lm
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
##1
df = pd.DataFrame({
    'Caffeine(mg)': [
        94.2, 93.7, 95.5, 93.9, 94.0, 95.2, 94.7, 93.5, 92.8, 94.4,
        93.8, 94.6, 93.3, 95.1, 94.3, 94.9, 93.9, 94.8, 95.0, 94.2,
        93.7, 94.4, 95.1, 94.0, 93.6
    ]
})

#1
print(df['Caffeine(mg)'].mean())
#2
_, pvalue = stats.shapiro(df['Caffeine(mg)'])
print(pvalue) # => 귀무가설 채택, df는 정규성을 만족한다.
#3
stat, pvalue = stats.ttest_1samp(df['Caffeine(mg)'], 95, alternative='less')
print(stat)
#4
print(pvalue)
#5
print('기각')
```

```python
##2
df = pd.DataFrame({
    '충전기': ['New'] * 10 + ['Old'] * 10,
    '충전시간': [
        1.5, 1.6, 1.4, 1.7, 1.5, 1.6, 1.7, 1.4, 1.6, 1.5,
        1.7, 1.8, 1.7, 1.9, 1.8, 1.7, 1.8, 1.9, 1.7, 1.6
    ]
})
#1
news = df['충전기'] == 'New'
olds = df['충전기'] == 'Old'
stat, pvalue = stats.ttest_ind(df[news]['충전시간'], df[olds]['충전시간'], alternative='less')
print(stat)
#2
print(pvalue)
#3
print('기각')
```

```python
##3
df = pd.DataFrame({
    'User': list(range(1, 11)),
    '기존방법': [60.4, 60.7, 60.5, 60.3, 60.8, 60.6, 60.2, 60.5, 60.7, 60.4],
    '새로운방법': [59.8, 60.2, 60.1, 59.9, 59.7, 58.4, 57.0, 60.3, 59.6, 59.8]
})
#1
df['ud'] = df['새로운방법'] - df['기존방법']
print(df['ud'].mean())
#2
stat, pvalue = stats.ttest_rel(df['ud'], 0, alternative='less')
print(stat)
#3
print(pvalue)
#4
print('기각')
```

```python
##4
df = pd.read_csv('dataset/part3/ch6/math.csv')
#1
a = df[df['groups'] == 'group_A']['scores']
_, pvalue = stats.shapiro(a)
print(pvalue)
b = df[df['groups'] == 'group_B']['scores']
_, pvalue = stats.shapiro(b)
print(pvalue)
c = df[df['groups'] == 'group_C']['scores']
_, pvalue = stats.shapiro(c)
print(pvalue)
d = df[df['groups'] == 'group_D']['scores']
_, pvalue = stats.shapiro(d)
print(pvalue)
#2
_, pvalue = stats.levene(a, b, c, d)
print(pvalue)
#3
model = ols('scores ~ C(groups)', data=df).fit()
re = anova_lm(model)
print('기각')
#4
print(re.iloc[0]['df'])
#5
print(re.iloc[1]['df'])
#6
print(re.iloc[0]['sum_sq'])
#7
print(re.iloc[0]['mean_sq'])
#8
print(re.iloc[0]['F'])
#9
print(re.iloc[0]['PR(>F)'])
```

```python
##6
a = [550, 250, 100, 70, 30]
b = [int(1000*i) for i in [0.6, 0.25, 0.08, 0.05, 0.02]]
#1
print(a[4] / sum(a))
#2
stat, pvalue = stats.chisquare(a, b)
print(stat)
#3
print(pvalue)
#4
print('기각')
```

```python
##7
#1
df = pd.DataFrame([[50, 30], [60, 40]])
chi2 = stats.chi2_contingency(df)
print(chi2.statistic)
#2
print(chi2.pvalue)
#3
print('채택')
#4
df = pd.DataFrame({
        '캠프': ['빅분기']*80 + ['정처기']*100,
        '등록여부': ['등록']*50 + ['등록안함']*30 + ['등록']*60 + ['등록안함']*40
})
df = pd.crosstab(df['캠프'], df['등록여부'])
chi2 = stats.chi2_contingency(df)
print(chi2.statistic)
#5
print(chi2.pvalue)
#6
print('채택')
```

```python
#8
df = pd.DataFrame({
    '할인율': [28, 24, 13, 0, 27, 30, 10, 16, 6, 5, 7, 11, 11, 30, 25,
            4, 7, 24, 19, 21, 6, 10, 26, 13, 15, 6, 12, 6, 20, 2],
    '온도': [15, 34, 15, 22, 29, 30, 14, 17, 28, 29, 19, 19, 34, 10,
           29, 28, 12, 25, 32, 28, 22, 16, 30, 11, 16, 18, 16, 33, 12, 22],
    '광고비': [342, 666, 224, 764, 148, 499, 711, 596, 797, 484, 986, 347, 146, 362, 642,
            591, 846, 260, 560, 941, 469, 309, 730, 305, 892, 147, 887, 526, 525, 884],
    '주문량': [635, 958, 525, 25, 607, 872, 858, 732, 1082, 863, 904, 686, 699, 615, 893,
            830, 856, 679, 918, 951, 789, 583, 988, 631, 866, 549, 910, 946, 647, 943]
})
#1
print(round(df['할인율'].corr(df['온도']), 2))
#2
model = ols('주문량 ~ 할인율 + 온도 + 광고비', data=df).fit()
print(round(model.rsquared, 2))
#3
print(round(model.params, 4))
#4
print(round(model.params['Intercept'], 4))
#5
print(round(model.pvalues['온도'], 4))
#6
x = pd.DataFrame({
    '할인율': [10],
    '온도': [20],
    '광고비': [500]
})
pred = model.predict(x)[0]
print(pred)
#7
df['잔차'] = df['주문량'] - model.predict(df)
print((df['잔차']**2).sum())
#8
print((df['잔차']**2).mean())
#9
print(list(model.conf_int(alpha=0.1).iloc[2]))
#10
x = pd.DataFrame({
    '할인율': [15],
    '온도': [25],
    '광고비': [300]
})
pred = model.get_prediction(x)
pred = pred.summary_frame(alpha=0.1)
print(list(pred.iloc[0, 2:4]))
print(list(pred.iloc[0, 4:6]))
#11
pvalue = model.pvalues['광고비']
print('기각')
```
