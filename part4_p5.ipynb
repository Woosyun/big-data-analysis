```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import root_mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
#1.1
df = pd.read_csv('dataset/part4/ch5/data5-1.csv')

cond1 = df['종량제봉투종류'] == '규격봉투'
cond2 = df['종량제봉투용도'] == '음식물쓰레기'
cond3 = df['2ℓ가격'] != 0
result = df[cond1 & cond2 & cond3]['2ℓ가격'].mean()
int(round(result, 0))
```

```python
#1.2
df = pd.read_csv('dataset/part4/ch5/data5-2.csv')
df.head()

df['bmi'] = df['Weight'] / (df['Height'] / 100)**2

normals = (df['bmi'] >= 18.5) & (df['bmi'] < 23)
normals = sum(normals)
dangers = (df['bmi'] >= 23) & (df['bmi'] < 25)
dangers = sum(dangers)
normals, dangers
abs(normals - dangers)
```

```python
#1.3
df = pd.read_csv('dataset/part4/ch5/data5-3.csv')

df['순전입학생수'] = df['전입학생수(계)'] - df['전출학생수(계)']
index = df.sort_values('순전입학생수', ascending=False).index[0]
int(df.loc[index, '전체학생수(계)'])
```

```python
#2
train = pd.read_csv('dataset/part4/ch5/train.csv')
test = pd.read_csv('dataset/part4/ch5/test.csv')

y_train = train.pop('price')

cc = pd.concat([train, test], axis=0)
cols = cc.select_dtypes(include='O').columns
for col in cols:
    le = LabelEncoder()
    cc[col] = le.fit_transform(cc[col])

train = cc.iloc[:len(train)].copy()
test = cc.iloc[len(train):].copy()

X_train, X_val, y_train, y_val = train_test_split(train, y_train, test_size=0.2, random_state=0)

rf = RandomForestRegressor(random_state=0)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_val)

rmse = root_mean_squared_error(y_val, y_pred)
print(rmse)

pred = rf.predict(test)
result = pd.DataFrame({
    'pred': pred
})
result.head()
```
