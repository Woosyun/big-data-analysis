```python
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
df = pd.read_csv('dataset/part3/ch2/fertilizer.csv')
model = ols('성장~C(비료)', df).fit()
anova_lm(model)
```

```python
df = pd.read_csv('dataset/part3/ch2/tree.csv')
model = ols('성장률~C(나무)+C(비료)+C(나무):C(비료)+C(비료):C(나무)', data=df).fit()
anova_lm(model)
```
