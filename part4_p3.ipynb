```python
#2
train = pd.read_csv('dataset/part4/ch3/train.csv')
test = pd.read_csv('dataset/part4/ch3/test.csv')

y_train = train.pop('TravelInsurance')

cc = pd.concat([train, test], axis=0)
cols = cc.select_dtypes(include='O').columns
for col in cols:
    le = LabelEncoder()
    cc[col] = le.fit_transform(cc[col])

train = cc.iloc[:len(train)].copy()
test = cc.iloc[len(train):].copy()

X_train, X_val, y_train, y_val = train_test_split(train, y_train, test_size=0.2, random_state=0)

rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)
y_pred_proba = rf.predict_proba(X_val)
roc_auc = roc_auc_score(y_val, y_pred_proba[:, 1])
print(roc_auc)

pred_proba = rf.predict_proba(test)
result = pd.DataFrame({
    'index': test.index,
    'y_pred': pred_proba[:, 1]
})
result
```
