```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
from scipy.stats import chisquare
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
#1.1
df = pd.read_csv('dataset/part4/ch6/data6-1-1.csv')
# df['diff'] = df['도착시간'] - df['출동시간']
df['도착시간'] = pd.to_datetime(df['도착시간'])
df['출동시간'] = pd.to_datetime(df['출동시간'])
df['diff'] = df['도착시간'] - df['출동시간']

result = df.groupby('소방서')['diff'].mean().sort_values(ascending=False).iloc[0].total_seconds()
result = round(result / 60)
result
```

```python
#1.2
df = pd.read_csv('dataset/part4/ch6/data6-1-2.csv')
df['per'] = df.iloc[:, 2:].sum(numeric_only=True, axis=1) / df['교사수']
result = df.sort_values('per', ascending=False).iloc[0, 1]
print(result)
```

```python
#1.3
df = pd.read_csv('dataset/part4/ch6/data6-1-3.csv')
df['범죄'] = df.iloc[:, 1:-1].sum(numeric_only=True, axis=1)
df.head()
df['year'] = df['날짜'].str[:4]
re = df.groupby('year')['범죄'].mean()
round(re.max())
```

```python
#2
train = pd.read_csv('dataset/part4/ch6/energy_train.csv')
test = pd.read_csv('dataset/part4/ch6/energy_test.csv')

y_train = train.pop('Heat_Load')

cc = pd.concat([train, test], axis=0)
cols = cc.select_dtypes(include='O').columns
for col in cols:
    le = LabelEncoder()
    cc[col] = le.fit_transform(cc[col])

train = cc.iloc[:len(train)].copy()
test = cc.iloc[:len(test)].copy()

X_train, X_val, y_train, y_val = train_test_split(train, y_train, test_size=0.2, random_state=0)

rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_val)

f1 = f1_score(y_val, y_pred, average='macro')
print(f1)

pred = rf.predict(test)
result = pd.DataFrame({
    'pred': pred
})
result.head()
```
