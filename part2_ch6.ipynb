```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, f1_score, accuracy_score
from sklearn.preprocessing import LabelEncoder, RobustScaler
```

```python
train = pd.read_csv('dataset/part2/ch6/diabetes_train.csv')
test = pd.read_csv('dataset/part2/ch6/diabetes_test.csv')

y_train = train.pop('Outcome')

X_train, X_val, y_train, y_val = train_test_split(train, y_train, test_size=0.2, random_state=0)

rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)
y_pred_proba = rf.predict_proba(X_val)
roc_auc_proba = roc_auc_score(y_val, y_pred_proba[:, 1])
display(roc_auc_proba)

pred = rf.predict_proba(test)
pred = pd.DataFrame({'pred': pred[:, 1]})
pred.to_csv('result.csv', index=False)
```

```python
train = pd.read_csv('dataset/part2/ch6/hr_train.csv')
test = pd.read_csv('dataset/part2/ch6/hr_test.csv')

y_train = train.pop('target')

cols = train.select_dtypes(include='O').columns
# for col in cols:
#     train[col] = train[col].fillna(train[col].mode()[0])
#     test[col] = test[col].fillna(train[col].mode()[0])

cc = pd.concat([train, test], axis=0)
for col in cols:
    le = LabelEncoder()
    cc[col] = le.fit_transform(cc[col])

train = cc.iloc[:len(train)].copy()
test = cc.iloc[len(train):].copy()

scaler = RobustScaler()
train = scaler.fit_transform(train)
test = scaler.transform(test)

X_train, X_val, y_train, y_val = train_test_split(train, y_train, test_size=0.2, random_state=0)

rf = RandomForestClassifier(random_state=0, max_depth=7, n_estimators=200)
rf.fit(X_train, y_train)
y_pred_proba = rf.predict_proba(X_val)

roc_auc_proba = roc_auc_score(y_val, y_pred_proba[:, 1])
roc_auc_proba
```

```python
train = pd.read_csv('dataset/part2/ch6/creditcard_train.csv')
test = pd.read_csv('dataset/part2/ch6/creditcard_test.csv')

y_train = train.pop('STATUS')

# train['OCCUPATION_TYPE'] = train['OCCUPATION_TYPE'].fillna(train['OCCUPATION_TYPE'].mode()[0])

cc = pd.concat([train, test], axis=0)
cc = cc.drop('ID', axis=1)
cols = cc.select_dtypes(include='O').columns
le = LabelEncoder()
for col in cols:
    cc[col] = le.fit_transform(cc[col])

train = cc.iloc[:len(train)].copy()
test = cc.iloc[len(train):].copy()

X_train, X_val, y_train, y_val = train_test_split(train, y_train, test_size=0.2, random_state=0)

rf = RandomForestClassifier(random_state=0, n_estimators=300, class_weight='balanced')
rf.fit(X_train, y_train)
y_pred = rf.predict(X_val)
f1 = f1_score(y_val, y_pred)
f1
```
