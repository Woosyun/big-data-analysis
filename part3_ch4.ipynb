```python
import pandas as pd
from statsmodels.formula.api import ols
```

```python
df = pd.DataFrame({
    '키': [150, 160, 170, 175, 165],
    '몸무게': [42, 50, 70, 64, 56],
})
df.corr()
```

```python
df = pd.read_csv('dataset/part3/ch4/study.csv')
# df = pd.get_dummies(df)

model = ols('score~study_hours+material_type', data=df).fit()
model.summary()

data = pd.DataFrame({
        'study_hours': [45],
        'material_type': ['독학']
    })
pred = model.predict(data)
pred
```
