```python
import pandas as pd
```

```python
#1
df = pd.read_csv('dataset/part4/ch9/loan.csv')

#1
df['총대출액'] = df['신용대출'] + df['담보대출']
#2
re = df.groupby(['지역코드', '성별'])['총대출액'].sum().unstack()
#3
re['diff'] = abs(re[1] - re[2])
idx = re['diff'].idxmax()
idx
```

```python
df = pd.read_csv('dataset/part4/ch9/crime.csv')

df.groupby(['연도', '구분']).sum().reset_index()

df1 = df[df['구분'] == '검거건수'].reset_index(drop=True)
df2 = df[df['구분'] == '발생건수'].reset_index(drop=True)
df11 = df1.drop(['구분', '연도'], axis=1)
df22 = df2.drop(['구분', '연도'], axis=1)
df = df11 / df22
#2
# df.max(axis=0)
df['most'] = df.idxmax(axis=1)
result = 0
row = 0
for col in df['most']:
    result += df1.loc[row, col]
    row+=1
print(result)
```

```python
#1.3
df = pd.read_csv('dataset/part4/ch9/hr.csv')
#1
df['만족도'] = df['만족도'].fillna(df['만족도'].mean())
#2
g = df.groupby(['부서', '성과등급'])['근속연수'].transform('mean').astype(int)
g
df['근속연수'] = df['근속연수'].fillna(g)
#3
df['연봉/근속연수'] = df['연봉'] / df['근속연수']
re3 = df.nlargest(3, '연봉/근속연수').iloc[-1]['근속연수']
print(re3)
#4
df['연봉/만족도'] = df['연봉'] / df['만족도']
re4 = df.nlargest(2, '연봉/만족도').iloc[-1]['교육참가횟수']
print(re4)
#5
re3 + re4
```
