```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, root_mean_squared_error
```

```python
train = pd.read_csv('dataset/part2/ch7/score_train.csv')
test = pd.read_csv('dataset/part2/ch7/score_test.csv')

y_train = train.pop('Credit_Score')

cc = pd.concat([train, test], axis=0)
cols = cc.select_dtypes(include='O').columns
for col in cols:
    le = LabelEncoder()
    cc[col] = le.fit_transform(cc[col])

train = cc[:len(train)].copy()
test = cc[len(train):].copy()

X_train, X_val, y_train, y_val = train_test_split(train, y_train, test_size=0.2, random_state=0)

rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_val)
f1 = f1_score(y_val, y_pred, average='macro')
f1
```

```python
train = pd.read_csv('dataset/part2/ch7/score_train.csv')
test = pd.read_csv('dataset/part2/ch7/score_test.csv')

y_train = train.pop('Credit_Score')

scaler = StandardScaler()
cols = train.select_dtypes(include=['int', 'float']).columns
train[cols] = scaler.fit_transform(train[cols])
test[cols] = scaler.transform(test[cols])

train = pd.get_dummies(train)
test = pd.get_dummies(test)

X_train, X_val, y_train, y_val = train_test_split(train, y_train, test_size=0.2, random_state=0)

rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_val)
f1 = f1_score(y_val, y_pred, average='macro')
f1
```
