```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
```

```python
#1
df = pd.read_csv('dataset/part4/ch4/data4-1.csv')
diff = df['age'].quantile(0.75) - df['age'].quantile(0.25)
diff = int(abs(diff))
print(diff)
```

```python
#3
df = pd.read_csv('dataset/part4/ch4/data4-3.csv')
df['date_added'] = pd.to_datetime(df['date_added'])
cond1 = (df['date_added'].dt.year == 2018) & (df['date_added'].dt.month == 1)
cond2 = df['country'] == 'United Kingdom'
len(df[cond1 & cond2])
```

```python
#2.1
train = pd.read_csv('dataset/part4/ch4/train.csv')
test = pd.read_csv('dataset/part4/ch4/test.csv')

y_train = train.pop('Segmentation')

cc = pd.concat([train, test], axis=0)
cols = cc.select_dtypes(include='O').columns
for col in cols:
    le = LabelEncoder()
    cc[col] = le.fit_transform(cc[col])

train = cc.iloc[:len(train)].copy()
test = cc.iloc[len(train):].copy()

X_train, X_val, y_train, y_val = train_test_split(train, y_train, test_size=0.2, random_state=0)

rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_val)
f1 = f1_score(y_val, y_pred, average='macro')
print(f1)

pred = rf.predict(test)
result = pd.DataFrame({
    'ID': test['ID'],
    'Segmentation': pred
})
result.head()
```
