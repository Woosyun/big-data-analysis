```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from statsmodels.formula.api import logit, ols
from sklearn.metrics import accuracy_score
```

```python
#1.1
df = pd.read_csv('dataset/part4/ch7/student_assessment.csv')
df = df.dropna()
idx = df['id_assessment'].value_counts().idxmax()
df = df[df['id_assessment'] == idx]

scaler = StandardScaler()
df['score'] = scaler.fit_transform(df[['score']])
round(df['score'].max(), 3)
```

```python
#1.2
df = pd.read_csv('dataset/part4/ch7/stock_market.csv')

idx = df.corr()['close'].loc['DE1':'DE77'].abs().idxmax()
result = df.mean()[idx]
round(result, 4)
```

```python
#3
df = pd.read_csv('dataset/part4/ch7/clam.csv')

train = df.iloc[:210].copy()
test = df.iloc[210:].copy()

#1
model1 = logit('gender ~ weight', data=df).fit()
odds = np.exp(model1.params['weight'])
print(round(odds, 4))
#2
model2 = logit('gender ~ age + length + diameter + height + weight', data=df).fit()
print(round(-2 * model2.llf, 2))
#3
pred = model1.predict(test)
pred = (pred > 0.5).astype(int)
accuracy = accuracy_score(pred, test['gender'])
print(round(1-accuracy, 3))
```

```python
#3.2
df = pd.read_csv('dataset/part4/ch7/system_cpu.csv')

#1
re1 = df.corr()['ERP'].iloc[1:].abs().max()
print(round(re1, 3))
#2
cond = df['CPU'] < 100
df = df[cond]
model = ols('ERP ~ Feature1 + Feature2 + Feature3 + CPU', data=df).fit()
print(round(model.rsquared, 3))
#3
re3 = model.pvalues.iloc[1:].max()
print(round(re3, 3)) 
```
