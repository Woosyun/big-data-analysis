```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, root_mean_squared_error, r2_score
```

```python
train = pd.read_csv('dataset/part2/ch8/flight_train.csv')
test = pd.read_csv('dataset/part2/ch8/flight_test.csv')

y_train = train.pop('price')

cc = pd.concat([train, test], axis=0)
cols = cc.select_dtypes(include='O').columns
for col in cols:
    le = LabelEncoder()
    cc[col] = le.fit_transform(cc[col])
train = cc.iloc[:len(train)].copy()
test = cc.iloc[len(train):].copy()

X_train, X_val, y_train, y_val = train_test_split(train, y_train, test_size=0.2, random_state=0)

rf = RandomForestRegressor(random_state=0)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_val)

rmse = root_mean_squared_error(y_val, y_pred)
rmse
```

```python
train = pd.read_csv('dataset/part2/ch8/flight_train.csv')
test = pd.read_csv('dataset/part2/ch8/flight_test.csv')

y_train = train.pop('price')

scaler = StandardScaler()
cols = train.select_dtypes(include='number').columns
train[cols] = scaler.fit_transform(train[cols])
test[cols] = scaler.transform(test[cols])

# train = pd.get_dummies(train)
# test = pd.get_dummies(test)

cc = pd.concat([train, test], axis=0)
cols = cc.select_dtypes(include='O').columns
for col in cols:
    le = LabelEncoder()
    cc[col] = le.fit_transform(cc[col])
train = cc.iloc[:len(train)].copy()
test = cc.iloc[len(train):].copy()

X_train, X_val, y_train, y_val = train_test_split(train, y_train, test_size=0.2, random_state=0)

rf = RandomForestRegressor(random_state=0, max_depth=20, n_estimators=200)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_val)

rmse = root_mean_squared_error(y_val, y_pred)
rmse
```

```python
train = pd.read_csv('dataset/part2/ch8/laptop_train.csv')
test = pd.read_csv('dataset/part2/ch8/laptop_test.csv')

y_train = train.pop('Price')

#drop
cols = ['Model', 'Series']
for col in cols:
    train = train.drop(col, axis=1)
    test = test.drop(col, axis=1)

#fill mode
cols = ['Hard_Disk_Capacity']
for col in cols:
    train[col] = train[col].fillna(train[col].mode()[0])
    test[col] = test[col].fillna(train[col].mode()[0])

#fill X
cols = ['Processor', 'Processor_Gen', 'OS']
for col in cols:
    train[col] = train[col].fillna('X')
    test[col] = test[col].fillna('X')

#fill mean
cols = ['RAM']
for col in cols:
    train[col] = train[col].fillna(train[col].mean())
    test[col] = test[col].fillna(train[col].mean())

#scale
# scaler = StandardScaler()
# cols = ['RAM', 'Rating']
# train[cols] = scaler.fit_transform(train[cols])
# test[cols] = scaler.transform(test[cols])

cc = pd.concat([train, test], axis=0)
cols = cc.select_dtypes(include='O').columns
for col in cols:
    le = LabelEncoder()
    cc[col] = le.fit_transform(cc[col])
train = cc.iloc[:len(train)].copy()
test = cc.iloc[len(train):].copy()

X_train, X_val, y_train, y_val = train_test_split(train, y_train, test_size=0.2, random_state=0)

rf = RandomForestRegressor(random_state=0)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_val)

r2 = r2_score(y_val, y_pred)
r2 
# base = 0.7
# 1. after delete ['Brand', 'Series] = 0.7
#   1.1 max_depth=20, n_estimators = 200 = 0.746
# 2. delete ['Model', 'Series'] = 0.788
```
